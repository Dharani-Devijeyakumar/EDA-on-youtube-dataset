### 5.Convert datatype of column published_at from object to pandas datetime.

```python
# to_dataetime 
youtube_data['published_at'] = pd.to_datetime(youtube_data['published_at'])
#to check whether it changes to datetime datatype.
print(youtube_data['published_at'].dtypes)

youtube_data['published_at']
```

### 6. Create a new column as 'published_month' using the column published_at (display the months only)

```python
# strftime('%m') to get the month (it returns the month with a number)
youtube_data['published_month'] = youtube_data['published_at'].dt.strftime('%m')
print(youtube_data['published_month'])
youtube_data.head()
```

### 7. Replace the numbers in the column published_month as names of the months i,e., 1 as 'Jan', 2 as 'Feb'and so on.....

```python
# strftime('%b') to get name of the month.
youtube_data['published_month'] = youtube_data['published_at'].dt.strftime('%b')
print(youtube_data['published_month'])
youtube_data.head()
```

### 8. Find the number of videos published each month and arrange the months in a decreasing order based on the video count.


```python
youtube_data.groupby('published_month').agg({'video_id':'count'}).sort_values(by='video_id',ascending=False)
```

### 9. Find the count of unique video_id, channel_id and channel_title.

### 10. Find the top10 channel names having the highest number of videos in the dataset and the bottom10 having lowest number of videos.

```python
# get top 10 and bottom 10 using sort_values
# also get the top and bottom values using nlargest and n smallest functions is done below.
df=youtube_data.groupby('channel_title').agg({'video_id':'count'}).sort_values(by='video_id',ascending=False)
print('Top 10 channels :' ,df.head(10))
print('\nBottom 10 channels :' ,df.tail(10))
```

```python
# to get top 10 channel names having the highest number of videos
top10 = youtube_data.groupby('channel_title')['video_id'].count().nlargest(10).to_frame()
top10
```

```python
# to get bottom 10 channel names having the lowest number of videos
bottom10 = youtube_data.groupby('channel_title')['video_id'].count().nsmallest(10).to_frame()
bottom10
```

```python
pd.concat([top10,bottom10],axis=0)
```

### 13. Does the number of views have any effect on how many people disliked the video? Support your answer with a metric and a plot.

```python
# for bivariate analysis I prefer to use scatter plot
plt.figure(figsize=(4,4))
plt.scatter(x=youtube_data['view_count'],y=youtube_data['dislikes'],c='red',alpha=0.7)
plt.xlabel('view_count')
plt.ylabel('dislikes')
plt.title('Views vs Dislikes')
```
